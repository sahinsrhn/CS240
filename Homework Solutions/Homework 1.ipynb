```python
hist = thinkstats2.Hist(resp.fmarno , label="Number of Mariage")
thinkplot.Hist(hist)
thinkplot.Config(xlabel='value of your data', ylabel='frequency')
```

```python
#Q5
maried = resp[resp.fmarno > 0]
neverMaried = resp[resp.fmarno == 0]

maried_hist = thinkstats2.Hist(maried.totincr, label='Maried')
neverMaried_hist = thinkstats2.Hist(neverMaried.totincr, label='Never Maried')
thinkplot.PrePlot(2)
thinkplot.Hist(maried_hist, align='right', width=0.5)
thinkplot.Hist(neverMaried_hist, align='left', width=0.5)
thinkplot.Config(xlabel='Total Income', ylabel='Frequency')
```

```python
#Q6
married_statistic = "For the Maried Respondent Minumum Income: {}, Maximum Income: {}, \
Standart Deviation: {}, Variance: {}, and Mean: {}"
married_statistic = married_statistic.format(maried.totincr.min(),maried.totincr.max(), \
                                             maried.totincr.std(),maried.totincr.var(),maried.totincr.mean())
married_statistic
```

```python
neverMarried_statistic = "For the Maried Respondent Minumum Income: {}, Maximum Income: {},\
Standart Deviation: {}, Variance: {}, and Mean: {}"
neverMarried_statistic = neverMarried_statistic.format(neverMaried.totincr.min(),neverMaried.totincr.max(), \
                                                       neverMaried.totincr.std(),neverMaried.totincr.var(), \
                                                       neverMaried.totincr.mean())
neverMarried_statistic
```
